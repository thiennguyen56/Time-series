```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.model_selection import train_test_split
```

```python
raw_data = pd.read_csv('COVID-19-geographic-disbtribution-worldwide-2020-04-30.csv')
raw_data['dateRep'] = pd.to_datetime(raw_data['dateRep'])
raw_data.head()
```

```python
vn_data = raw_data[raw_data['geoId'] == 'VN']
vn_data.to_csv('Covid-19_VietNam.csv')
len(vn_data)
```

```python
us_data = raw_data[raw_data['geoId'] == 'US']
us_data.to_csv('Covid-19_US.csv')
len(us_data)
it_data = raw_data[raw_data['geoId'] == 'IT']
it_data.to_csv('Covid-19_IT.csv')
print(len(it_data))
fr_data = raw_data[raw_data['geoId'] == 'FR']
len(fr_data)
fr_data.to_csv('Covid-19_FR.csv')
```

```python
vn_data['dateRep'] = pd.to_datetime(vn_data['dateRep'])
```

```python
# EDA 

plt.figure(figsize=(16,9))
x = vn_data['dateRep']
y1 = vn_data['cases']
y2 = vn_data['deaths']
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('Coronaviruses are happening in Vietnam')
plt.xlabel('Day')
plt.ylabel('A mount of Cases and Deaths')
plt.legend(['Cases', 'Deaths'])
plt.show()
```
